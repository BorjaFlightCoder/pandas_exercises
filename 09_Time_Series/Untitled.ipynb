```python
def generar_dataframe(n_fechas = 24, start = "2018-01-01", freq="M"):

    meses = pd.date_range(start, periods=n_fechas, freq=freq)

    vuelos = np.random.normal(loc=370, scale=10, size=n_fechas).round()

    ticket_medio = np.random.normal(loc=70, scale=20, size=n_fechas)

    ventas = np.random.normal(loc=370, scale=10, size=n_fechas)

    pasajeros_vuelo = np.random.normal(loc=155, scale=5, size=n_fechas).round()

    df = pd.DataFrame({'mes': meses,
                 'vuelos': vuelos,
                 'precio_ticket': ticket_medio,
                      'pasajeros_x_vuelo': pasajeros_vuelo})

    df['ingresos'] = df.vuelos * df.precio_ticket
    df['total_pasajeros'] = df.pasajeros_x_vuelo * df.vuelos

    return df
```
